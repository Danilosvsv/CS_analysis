```python
def list_webpages(url):
    """
        entry: fornece uma url para a função
        lista_urls: retorna uma lista contendo o acesso de cada webpage
    """
    link = url
    f = requests.get(link)
    lista_urls = []
    
    # calculando o total de páginas a partir do html da primeira página
    
    total_count = 285
    num_pages = round(total_count/10)
    
    for i in range(3): # substituir por num_pages para ter acesso a todas subpaginas
        subpages = "#p"+str(i)+"_popular_desc"
        html_subpages = requests.get(link+subpages)
        lista_urls.append(html_subpages)
        
    return lista_urls
```
